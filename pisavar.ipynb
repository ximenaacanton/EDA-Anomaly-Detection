```python
from scipy.stats import zscore

# Convertir a tipo num√©rico si no lo est√°
df['value'] = pd.to_numeric(df['value'], errors='coerce')

# Eliminar filas con NaN generados por la conversi√≥n
df = df.dropna(subset=['value'])

# Calcular Z-score
z_scores = zscore(df['value'])

# Detectar anomal√≠as donde el Z-score es mayor que 3 o menor que -3
anomalies_z = df[abs(z_scores) > 3]

print(anomalies_z)
```

```python
df = pd.read_csv('/Users/ximenacanton/Documents/TEC/4 SEM/ACD/pisa/df_CONTIFORM_MMA_CONTIFORM_MMA1_WS_Tot_Rej_0.csv')
print(df.head())  # Muestra las primeras filas del dataframe
print(df.info())  # Informaci√≥n sobre el dataframe, tipos de columnas
```

```python
#EDA
print(df.describe())  # Estad√≠sticas generales
print(df.info())  # Informaci√≥n de tipos de datos y valores nulos
print(df.isnull().sum())  # Revisi√≥n de valores nulos

#Distribuci√≥n
sns.histplot(df["value"], bins=30, kde=True,color="pink")
plt.title("Distribuci√≥n de cvalue")
plt.show()



#Outliers
sns.boxplot(x=df["value"],color="pink")
plt.show()

#Tendencias temporales
df.plot(x="user_ts", y="value", figsize=(10,5),color="pink")
plt.show()

```

```python
# Estad√≠sticas descriptivas avanzadas
desc_stats = df['value'].describe()
skewness = df['value'].skew()  # Asimetr√≠a
kurtosis = df['value'].kurtosis()  # Curtosis

print("Estad√≠sticas descriptivas:\n", desc_stats)
print("Asimetr√≠a:", skewness)
print("Curtosis:", kurtosis)
```

```python
# Gr√°fico de densidad
plt.figure(figsize=(10, 6))
sns.kdeplot(df['value'], shade=True)
plt.title('Distribuci√≥n de "value" con Gr√°fico de Densidad')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```

```python
import scipy.stats as stats

# QQ plot para comprobar normalidad
plt.figure(figsize=(10, 6))
stats.probplot(df['value'], dist="norm", plot=plt)
plt.title('QQ Plot de "value"')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aseg√∫rate de que 'user_ts' sea de tipo datetime
df['user_ts'] = pd.to_datetime(df['user_ts'])

# Graficar la variable 'value' a lo largo del tiempo
plt.figure(figsize=(10, 6))
plt.plot(df['user_ts'], df['value'], marker='o', color='pink')
plt.title('Tendencia de "value" a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
Q1 = df['value'].quantile(0.25)
Q3 = df['value'].quantile(0.75)
IQR = Q3 - Q1

# Definir los l√≠mites de los outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
anomalies = df[(df['value'] < lower_bound) | (df['value'] > upper_bound)]

print("Anomal√≠as detectadas:")
print(anomalies)
```
